```python
from scipy.stats import pearsonr
import pandas as pd

target = 'HeartDisease'

numerical_features = [
    'Age',
    'RestingBP',
    'Cholesterol',
    'MaxHR',
    'Oldpeak',
    'Age_Oldpeak'
]

pearson_results = []

for feature in numerical_features:
    
    corr, _ = pearsonr(
        df[feature],
        df[target]
    )
    
    pearson_results.append(
        [feature, corr]
    )

pearson_df = pd.DataFrame(
    pearson_results,
    columns=[
        'Feature',
        'Pearson_Correlation'
    ]
)

pearson_df = pearson_df.sort_values(
    by='Pearson_Correlation',
    key=abs,
    ascending=False
)

print(pearson_df)
```

```python
from scipy.stats import chi2_contingency
import pandas as pd

target_col = 'HeartDisease'

categorical_features = [
    'Sex',
    'FastingBS',
    'ExerciseAngina',
    'RestingECG_Normal',
    'RestingECG_ST',
    'ChestPainType_ATA',
    'ChestPainType_NAP',
    'ChestPainType_TA',
    'ST_Slope_Flat',
    'ST_Slope_Up',
    'Age_Group_Young',
    'Age_Group_Middle_Aged',
    'Age_Group_Senior',
    'High_BP'
]

alpha = 0.05

chi_results = []

for feature in categorical_features:

    contingency_table = pd.crosstab(
        df[feature],
        df[target_col]
    )

    chi_score, p_value, _, _ = chi2_contingency(
        contingency_table
    )

    decision = "Keep" if p_value < alpha else "Drop"

    chi_results.append([
        feature,
        chi_score,
        p_value,
        decision
    ])

chi_df = pd.DataFrame(
    chi_results,
    columns=[
        'Feature',
        'Chi_Square_Score',
        'P_Value',
        'Decision'
    ]
)

chi_df = chi_df.sort_values(
    by='Chi_Square_Score',
    ascending=False
)

print(chi_df)
```

```python
selected_features = [
    'ST_Slope_Up',
    'ST_Slope_Flat',
    'ExerciseAngina',
    'ChestPainType_ATA',
    'Sex',
    'FastingBS',
    'ChestPainType_NAP',
    'Age_Group_Senior',
    'Age_Group_Young',
    'High_BP',
    'RestingECG_ST',
    'RestingECG_Normal',
    'Age_Group_Middle_Aged',
    'MaxHR',
    'Age_Oldpeak',
    'Oldpeak',
    'Age'
]
df = df[selected_features +['HeartDisease']]
df
```

```python
# now train the multiple machine learning supervised learning models and select best:
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix,f1_score,accuracy_score

X = df.drop('HeartDisease',axis = 1)
y= df['HeartDisease']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

```

```python
models = {
    'logistic': LogisticRegression(),
    'svc': SVC(),
    'knn': KNeighborsClassifier(n_neighbors=5),
    'naive': GaussianNB(),
    'decision': DecisionTreeClassifier()
}

results = []

for name, model in models.items():

    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    acc = accuracy_score(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)
    f1_sc = f1_score (y_test, y_pred)

    results.append({
        'model': name,
        'accuracy': acc,
        'confusion': cm,
        'f1': f1_sc
    })

results
```
